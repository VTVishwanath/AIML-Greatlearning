## Lists in Python

Sequences in Python are data structures that hold objects in an ordered array. Now, we will work on Lists, the most common sequence data types in Python.

```python
#Example 
l1 = ['learning', "Python", 'is fun?', True]
print(l1)
```

List can also be created by using list() function.

### Question 4
Number List

Find Range and Mean of l3.

`l3 = [2,4,6,8]`

Hint: use **`len()`, `sum()`, `min()`, `max()`** functions
 
If you want to use standard functions like mean & range, you have to import them from numpy else you can calculate them the traditional way using formulas

```python
import numpy as np
l3 = [2,4,6,8]
print('List l3 is: ' + str(l3))
Mean = sum(l3)/len(l3)
Range = max(l3) - min(l3)
print('Mean of the list l3 is: ' + str(Mean))
print('Range of the list l3 is: ' + str(Range))
```

### Question 5

Hint: use **`+`** operator to add multiple elements in the array and **`count()`** function to print the occurences

### Question 6
Sorting and Reversing a list

sort and print l3 in ascending and descending order sequentially (given above)

Hint: use **`sort()`** function

```python
l3.sort()
print('After sorting the list l3 in ascending order: ' + str(l3))
l3.sort(reverse = True)
print('After sorting the list l3 in descending order: ' + str(l3))
l3.reverse()
print('After reversing the list l3: ' + str(l3))
```

# Numpy

We have seen python basic data structures in our last section. They are great but lack specialized features for data analysis. Like, adding rows, columns and operating on 2d matrices aren't readily available. So, we will use ***Numpy*** for such functions.



```python
import numpy as np
```

Numpy operates on *nd* arrays. These are similar to lists but contains homogenous elements but easier to store 2-d data.

```python
l1 = [1,2,3,4]
nd1 = np.array(l1)
print(nd1)

l2 = [5,6,7,8]
nd2 = np.array([l1,l2])
print(nd2)
```

Some functions on np.array()

```python
print(nd2.shape)

print(nd2.size)

print(nd2.dtype)
```

## Measures of Data

Descriptive Statistcs are measures that summarize important features of data. Producing these descriptive statistics is a common first step to take after cleaning and preparing a dataset for further analysis. Examples are like calculating Mean or Median. In this lab, we will be looking into some of these functions and explore several new ones.

#### Measures of Central Tendency

Whenever you measure things of the same kind, a fairly large number of such measurements will tend to cluster around the middle value. Such a value is called a measure of "Central Tendency". The other terms that are used synonymously are "Measures of Location", or "Statistical Averages".

## Measures of Dispersion

Measures of Dispersion (Spread) are statistics that describe how data varies. Measure of dispersion gives us the sense of how much the data tends to diverge from the central tendency.

### Question 5

What is the range of `mpg` in the given dataset?

**Hint:** Range = max - min

```python
Range = max(mtcars['mpg'])-min(mtcars['mpg'])
print(Range)
```

### Question 6

Calculate the 25th quantile (lower quantile) for `mpg` in the given dataset.

**Hint:** Use **.quantile()**

```python
mtcars['mpg'].quantile(.25)
```

### Question 7 quantileulate the Inter-Quartile Range (IQR) for `mpg` in the given dataset.

Interquartile range (IQR) is another common measure of spread. IQR is the distance between the 3rd quartile and the 1st quartile.

**Hint:** IQR = Q3 - Q1

```python
IQR = mtcars['mpg'].quantile(.75) - mtcars['mpg'].quantile(.25)
print(IQR)
```

Variance and Standard Deviation are two other common measure of dispersion.

**Hint:** Use .var() and .std() and .mad()

```python
mtcars['mpg'].std()
```
